```python
import pandas as pd
```

```python
datos_rss = pd.read_csv (r'TranformacionBases_Java/Rss_adap_multi.csv', engine='python', encoding = "ISO-8859-1", sep=';')
datos_twitter = pd.read_csv (r'TranformacionBases_Java/Twitter_adap_multi.csv', engine='python', encoding = "ISO-8859-1", sep=';')
```

```python
datos_rss
```

```python
datos_rss_fin=datos_rss[datos_rss['polaridad']!=0][['fuenteRss','date_adj','polaridad']]
```

```python
datos_rss_fin=datos_rss_fin.groupby(['fuenteRss','date_adj']).mean().reset_index()
datos_rss_fin.columns=['fuente','date','polaridad']
datos_rss_fin['tipo']='rss'
```

```python
datos_rss_fin
```

```python
datos_twitter
```

```python
datos_twitter_fin=datos_twitter[datos_twitter['polaridad']!=0][['id_fuente','date','polaridad']]
```

```python
datos_twitter_fin=datos_twitter_fin.groupby(['id_fuente','date']).mean().reset_index()
datos_twitter_fin.columns=['fuente','date','polaridad']
datos_twitter_fin['tipo']='twitter'
```

```python
datos_twitter_fin
```

```python
datos_base=datos_rss_fin.append(datos_twitter_fin)
```

```python
datos_base
```

```python
maestro_fuentes= datos_base[['fuente','tipo']].drop_duplicates()
```

```python
maestro_fuentes.reset_index(drop=True, inplace=True)
```

```python
maestro_fuentes
```

```python
maestro_fuentes.to_csv('TranformacionBases_Java/maestro_fuentes.csv', index=False, sep=';')
```
