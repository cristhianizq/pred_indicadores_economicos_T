```python
Tweets_resultados=tweeter_df_base[['id_fuente']].value_counts().to_frame()
```

```python
corr_Tweets=tweeter_df_base.groupby('id_fuente')[['SentTesis','Diferencia']].corr()
```

```python
corr_Tweets.reset_index(inplace=True)
corr_Tweets=corr_Tweets[corr_Tweets['level_1']=='Diferencia']
corr_Tweets.drop(columns=['Diferencia','level_1'], inplace=True)
```

```python
res_Tweets_resultados=Tweets_resultados.merge(corr_Tweets, left_on='id_fuente', right_on='id_fuente', how='left')
res_Tweets_resultados.columns=['id_fuente','cuenta','corr']
```

```python
res_Tweets_resultados.sort_values('corr')
```

```python
corr_Tweets=tweeter_df_base.groupby('id_fuente')[['SentTesis','Diferencia']].corr(method='pearson')
corr_Tweets.reset_index(inplace=True)
corr_Tweets=corr_Tweets[corr_Tweets['level_1']=='Diferencia']
corr_Tweets.drop(columns=['Diferencia','level_1'], inplace=True)
```

```python
res_Tweets_resultados=res_Tweets_resultados.merge(corr_Tweets, left_on='id_fuente', right_on='id_fuente', how='left')
res_Tweets_resultados.columns=['id_fuente','cuenta','corr','corrPearson']
```

```python
res_Tweets_resultados
```

```python
corr_Tweets=tweeter_df_base.groupby('id_fuente')[['SentTesis','Diferencia']].corr(method='kendall')
corr_Tweets.reset_index(inplace=True)
corr_Tweets=corr_Tweets[corr_Tweets['level_1']=='Diferencia']
corr_Tweets.drop(columns=['Diferencia','level_1'], inplace=True)
res_Tweets_resultados=res_Tweets_resultados.merge(corr_Tweets, left_on='id_fuente', right_on='id_fuente', how='left')
res_Tweets_resultados.columns=['id_fuente','cuenta','corr','corrPearson','corrKendall']
res_Tweets_resultados
```

```python
corr_Tweets=tweeter_df_base.groupby('id_fuente')[['SentTesis','Diferencia']].corr(method='spearman')
corr_Tweets.reset_index(inplace=True)
corr_Tweets=corr_Tweets[corr_Tweets['level_1']=='Diferencia']
corr_Tweets.drop(columns=['Diferencia','level_1'], inplace=True)
res_Tweets_resultados=res_Tweets_resultados.merge(corr_Tweets, left_on='id_fuente', right_on='id_fuente', how='left')
res_Tweets_resultados.columns=['id_fuente','cuenta','corr','corrPearson','corrKendall','corrSpearman']
res_Tweets_resultados
```
