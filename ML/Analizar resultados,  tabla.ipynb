```python
ruta_result_fuente = f"result_oro.pkl"
r_oro=load(ruta_result_fuente)
ruta_result_fuente = f"result_trm.pkl"
r_trm=load(ruta_result_fuente)
ruta_result_fuente = f"result_petroleo.pkl"
r_petroleo=load(ruta_result_fuente)
```

```python
df_rl=r_trm['rl'][['coef_abs']].merge(r_oro['rl'][['coef_abs']], left_index=True, right_index=True, how='outer')
df_rl=df_rl.merge(r_petroleo['rl'][['coef_abs']], left_index=True, right_index=True, how='outer')
df_rl.columns=['trm','oro','petroleo']
```

```python
df_rl
```

```python
# Resultados XGBOOST
```

```python
r_trm['xgb']
```

```python
df_xgb=r_trm['xgb'][['score']].merge(r_oro['xgb'][['score']], left_index=True, right_index=True, how='outer')
df_xgb=df_xgb.merge(r_petroleo['xgb'][['score']], left_index=True, right_index=True, how='outer')
df_xgb.columns=['trm','oro','petroleo']
```

```python
df_xgb
```

```python
r_trm['rf']
```

```python
r_oro['rf']
```

```python
df_rf=r_trm['rf'].merge(r_oro['rf'], left_on='fuente', right_on='fuente', how='outer')
df_rf=df_rf.merge(r_petroleo['rf'], left_on='fuente', right_on='fuente', how='outer')
df_rf.columns=['fuente','trm','oro','petroleo']
df_rf
```

```python
## Por ranking
```

```python
t1=r_trm['rl'].sort_values('coef_abs', ascending=False).reset_index().reset_index()[['index','level_0']].set_index('index')
t2=r_oro['rl'].sort_values('coef_abs', ascending=False).reset_index().reset_index()[['index','level_0']].set_index('index')
t3=r_petroleo['rl'].sort_values('coef_abs', ascending=False).reset_index().reset_index()[['index','level_0']].set_index('index')
df_rl_r=t1.merge(t2, left_index=True, right_index=True, how='outer')
df_rl_r=df_rl_r.merge(t3, left_index=True, right_index=True, how='outer')
df_rl_r.columns=['trm','oro','petroleo']
df_rl_r
```

```python
df_rl_r.sort_values(['trm','oro','petroleo']).to_csv('resultado_rl.csv')
```

```python
df_xgb_r.sort_values(['trm','oro','petroleo']).to_csv('resultado_xgb.csv')
```
